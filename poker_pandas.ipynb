```python
import pandas as pd
```

```python
# Load the data
# They'll be familiar with this
file1 = 'data1.csv'
data = pd.read_csv(file1)
data
```

```python
# Find the cases that match certain cards in a hand and return the corresponding rows
# They should be familiar with everything here except for the `.str.contains()` pattern
# All that does is converts the entry to a string if it's not already, then searches
# for a substring and returns True if it is found
def find_matches(df,card_list):
    matches = pd.Series(True, index=df.index)
    for card in card_list:
        matches = matches & df['Cards'].str.contains(card)
    return matches
    
card_list = ['Ah','Ks','Kh']
matches = find_matches(data,card_list)
data[matches]
```

```python
# Sum along a particular column
# They will be familiar with this
data['pair'].sum()
```

```python
# Sum along all columns
# Sum is familiar to them, but the `numeric_only` keyword is new (avoids concatenating strings as a "sum")
total = data.sum(axis=0, numeric_only=True)
total
```

```python
# Get percentages
percentages = total / total.sum() * 100
percentages
```
